````python
sp = spm.SentencePieceProcessor()
# 4/20
model_path = f"{model_prefix}.model"
#
sp.Load(model_path)

vocab = [(sp.IdToPiece(i), sp.GetScore(i)) for i in range(sp.GetPieceSize())]

tokenizer_obj = Tokenizer(Unigram(vocab, unk_id=2))
tokenizer_obj.pre_tokenizer = Metaspace()
tokenizer_obj.decoder = MetaspaceDecoder()

sentinel_tokens = [f"<extra_id_{i}>" for i in range(100)]

tokenizer = PreTrainedTokenizerFast(
    tokenizer_object=tokenizer_obj,
    eos_token="</s>",
    unk_token="<unk>",
    pad_token="<pad>",
    additional_special_tokens=sentinel_tokens,
)
````

````python
def evaluate_model(model, eval_dataloader, device):
    model.eval()
    eval_loss, num_batches = 0.0, 0
    with torch.no_grad():
        for batch in eval_dataloader:
            input_ids, attention_mask, labels = [x.to(device) for x in batch]
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            eval_loss += outputs.loss.item()
            num_batches += 1
    return round(eval_loss / num_batches, 5)


class JavaSpanDataset(Dataset):
    def __init__(self, hf_dataset, tokenizer, max_length=512, corruption_rate=0.15, mean_span_length=3):
        self.dataset = hf_dataset
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.corruption_rate = corruption_rate
        self.mean_span_length = mean_span_length

        # Pre-fetch sentinel IDs (assuming 100 available: <extra_id_0> to <extra_id_99>)
        sentinel_tokens = [f"<extra_id_{i}>" for i in range(100)]
        self.sentinel_ids = tokenizer.convert_tokens_to_ids(sentinel_tokens)

    def __len__(self):
        return len(self.dataset)

    def _generate_span_mask(self, n_tokens):
        """Generate a boolean mask array where True = corrupted, using T5-style span placement."""
        # Number of tokens to corrupt and number of spans
        n_corrupt = max(1, int(n_tokens * self.corruption_rate))
        n_spans = max(1, int(n_corrupt / self.mean_span_length))
        n_spans = min(n_spans, len(self.sentinel_ids))  # cap at available sentinels

        # Sample span lengths from a uniform distribution centered on mean_span_length,
        # then adjust so total corrupted tokens == n_corrupt
        span_lengths = [self.mean_span_length] * n_spans
        remaining = n_corrupt - sum(span_lengths)
        for i in range(abs(remaining)):
            if remaining > 0:
                span_lengths[i % n_spans] += 1
            elif remaining < 0:
                span_lengths[i % n_spans] = max(1, span_lengths[i % n_spans] - 1)

        # Place spans at random non-overlapping start positions
        # Think of it as distributing n_tokens - n_corrupt uncorrupted tokens into n_spans+1 gaps
        n_uncorrupted = n_tokens - sum(span_lengths)
        if n_uncorrupted < 0:
            # Edge case: more tokens to corrupt than available, truncate span lengths
            scale = n_tokens / sum(span_lengths)
            span_lengths = [max(1, int(l * scale)) for l in span_lengths]
            n_uncorrupted = n_tokens - sum(span_lengths)

        # Distribute uncorrupted tokens into (n_spans + 1) gaps
        n_gaps = n_spans + 1
        gaps = [0] * n_gaps
        for _ in range(n_uncorrupted):
            gaps[random.randint(0, n_gaps - 1)] += 1

        # Build the mask
        mask = [False] * n_tokens
        pos = 0
        for i, span_len in enumerate(span_lengths):
            pos += gaps[i]  # skip the gap of uncorrupted tokens
            for j in range(span_len):
                if pos + j < n_tokens:
                    mask[pos + j] = True
            pos += span_len

        return mask

    def __getitem__(self, idx):
        # 1. Get raw code and tokenize
        example = self.dataset[idx]
        code = example.get("whole_func_string") or example.get("func_code_string") or ""
        token_ids = self.tokenizer.encode(code, max_length=self.max_length, truncation=True)

        n_tokens = len(token_ids)
        mask = self._generate_span_mask(n_tokens)

        # 2. Single-pass: build input_ids and target_ids from the span mask
        input_ids = []
        target_ids = []
        sentinel_idx = 0
        is_in_span = False

        for i, token in enumerate(token_ids):
            if mask[i]:
                if not is_in_span:
                    # Starting a new masked span — insert sentinel
                    if sentinel_idx < len(self.sentinel_ids):
                        sentinel = self.sentinel_ids[sentinel_idx]
                        input_ids.append(sentinel)
                        target_ids.append(sentinel)
                        sentinel_idx += 1
                        is_in_span = True
                    else:
                        # Ran out of sentinels — keep token unmasked
                        input_ids.append(token)
                        continue
                # Add the actual token to targets
                target_ids.append(token)
            else:
                # Not masked
                is_in_span = False
                input_ids.append(token)

        target_ids.append(self.tokenizer.eos_token_id)

        # 3. Truncate (just in case sentinels pushed it over limit)
        input_ids = input_ids[:self.max_length]
        target_ids = target_ids[:self.max_length]

        # 4. Correct Attention Mask Calculation
        # The mask is 1 for actual tokens, BEFORE padding is added
        actual_length = len(input_ids)
        attention_mask = [1] * actual_length

        # 5. Pad Sequences
        def pad(ids, pad_val):
            return ids + [pad_val] * (self.max_length - len(ids))

        input_ids_padded = pad(input_ids, self.tokenizer.pad_token_id)
        attention_mask_padded = pad(attention_mask, 0)
        target_ids_padded = pad(target_ids, -100)

        return (
            torch.tensor(input_ids_padded, dtype=torch.long),
            torch.tensor(attention_mask_padded, dtype=torch.long),
            torch.tensor(target_ids_padded, dtype=torch.long)
        )

# 1. SETUP (Assuming tokenizer and model are initialized as per previous steps)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_a.to(device)

# 2. CREATE DATALOADERS
# Split the 50k into train/eval
split_data = methods.train_test_split(test_size=0.1)
train_ds = JavaSpanDataset(split_data["train"], tokenizer)
eval_ds = JavaSpanDataset(split_data["test"], tokenizer)

train_dataloader = DataLoader(train_ds, batch_size=8, shuffle=True, num_workers=4, pin_memory=True)
eval_dataloader = DataLoader(eval_ds, batch_size=8, num_workers=4, pin_memory=True)

# 3. OPTIMIZER & SCHEDULER
optimizer = optim.AdamW(model_a.parameters(), lr=5e-5)
epochs = 3
num_training_steps = epochs * len(train_dataloader)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=500, num_training_steps=num_training_steps)

# 4. TRAINING LOOP (Your code updated)
output_dir = "./java_t5_saved_model_pt"
os.makedirs(output_dir, exist_ok=True)
best_loss = float("inf")

#EPOCHS = 3

for epoch in range(epochs):
    model_a.train()
    tr_loss, tr_steps = 0.0, 0

    for step, batch in enumerate(train_dataloader):
        input_ids, attention_mask, labels = [x.to(device) for x in batch]

        outputs = model_a(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_a.parameters(), max_norm=1.0)

        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

        tr_loss += loss.item()
        tr_steps += 1

        if step % 100 == 0:
            print(f"Epoch {epoch} | Step {step} | Loss: {loss.item():.4f}")

    train_loss = round(tr_loss / tr_steps, 5)
    eval_loss = evaluate_model(model_a, eval_dataloader, device)

    print(f"Epoch {epoch} Complete. Train Loss: {train_loss} | Eval Loss: {eval_loss}")

    if eval_loss < best_loss:
        best_loss = eval_loss
        model_a.save_pretrained(output_dir)
        tokenizer.save_pretrained(output_dir)
````

# **Now, finetune both models one of which we pretrained**

````python
#Finetune the pretrained model

# 1. Load Dataset & Your Pre-trained Model
dataset = load_dataset("google/code_x_glue_cc_code_refinement", name="medium")

# Load the model and tokenizer you saved in your previous steps
model_path = "./java_t5_saved_model_pt"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)

# 2. Preprocessing: Buggy -> Fixed
max_length = 512

def preprocess_function(examples):
    # The input to the model is the buggy code.
    # T5 usually likes a task prefix, so we add "fix bug: "
    inputs = ["fix bug: " + code for code in examples["buggy"]]
    targets = examples["fixed"]

    # Tokenize the inputs
    model_inputs = tokenizer(inputs, max_length=max_length, truncation=True)

    # Tokenize the targets (the fixed code)
    labels = tokenizer(text_target=targets, max_length=max_length, truncation=True)

    # Hugging Face models expect the target token IDs to be under the key "labels"
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Apply the preprocessing to the whole dataset (this is very fast due to batched=True)
tokenized_datasets = dataset.map(preprocess_function, batched=True, remove_columns=dataset["train"].column_names)

# 3. Data Collator (The dynamic padder)
# This replaces your pad() function from earlier. It dynamically pads inputs and labels
# to the longest sequence in the current batch, which saves a ton of GPU memory.
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, label_pad_token_id=-100)

# Compute Metrics (CodeBLEU + sacrebleu during training)
bleu_metric = evaluate.load("sacrebleu")

def compute_metrics(eval_preds):
    preds, labels = eval_preds
    # Replace -100 (padding) with pad token id so the tokenizer can decode
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # sacrebleu (standard BLEU)
    sacrebleu_refs = [[label] for label in decoded_labels]
    bleu_result = bleu_metric.compute(predictions=decoded_preds, references=sacrebleu_refs)

    # CodeBLEU (code-aware: n-gram + syntax + dataflow)
    try:
        codebleu_result = calc_codebleu(
            references=[[ref] for ref in decoded_labels],
            predictions=decoded_preds,
            lang="java",
        )
        codebleu_score = codebleu_result["codebleu"] * 100  # scale to 0-100 to match BLEU
    except Exception as e:
        print(f"⚠️ CodeBLEU computation failed: {e}")
        codebleu_score = -1.0

    return {
        "bleu": bleu_result["score"],
        "codebleu": codebleu_score,
    }

# 4. Training Configuration
training_args = Seq2SeqTrainingArguments(
    output_dir="./t5_PRETRAINED_code_refinement_finetuned",
    evaluation_strategy="epoch",      # Evaluate at the end of every epoch
    save_strategy="epoch",            # Save at the end of every epoch (aligned with eval)
    learning_rate=3e-5,               # Fine-tuning needs a smaller LR than pre-training
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    save_total_limit=2,               # Only keep the 2 most recent checkpoints to save disk space
    num_train_epochs=5,
    predict_with_generate=True,       # Required for Seq2Seq generation during evaluation
    load_best_model_at_end=True,      # Restore the best checkpoint after training
    metric_for_best_model="codebleu", # Use CodeBLEU to determine the best checkpoint
    greater_is_better=True,           # Higher CodeBLEU = better
    fp16=True,                        # 🔥 Turns on Mixed Precision! Doubles speed on your T4 GPU
    dataloader_num_workers=4,         # Using the optimal workers we discussed
)

# 5. Initialize Trainer & Train
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

print("Starting Fine-tuning...")
trainer.train()

# 6. Save the final model
trainer.save_model("./t5_PRETRAINED_code_refinement_final")
````

````python
import os
import re
import json
import pickle
import numpy as np
from typing import List, Dict
from tqdm import tqdm

from datasets import load_dataset
from gensim.models import Word2Vec
from sklearn.neighbors import NearestNeighbors

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

from codebleu import calc_codebleu

print("All imports successful")
````

````python
# =====================================================================
# CELL 3: Load CodeXGLUE Code Refinement Dataset
# =====================================================================
# Same dataset used for fine-tuning in snippet.py
dataset = load_dataset("google/code_x_glue_cc_code_refinement", name="medium")

print(f"Train: {len(dataset['train'])} examples")
print(f"Validation: {len(dataset['validation'])} examples")
print(f"Test: {len(dataset['test'])} examples")

# Preview a sample
sample = dataset["train"][0]
print(f"\nBuggy:  {sample['buggy'][:200]}...")
print(f"Fixed:  {sample['fixed'][:200]}...")
````

````python
# =====================================================================
# CELL 4: Word2Vec Code Embedder
# =====================================================================
# Reused from the RAG lab — trains skip-gram on code tokens,
# then mean-pools token vectors to get one vector per code snippet.

def tokenize_code(text):
    """Simple whitespace tokenizer, lowercase."""
    return text.lower().split()


class CodeEmbedder:
    """Train Word2Vec on code, then mean-pool token vectors for sequence embeddings."""

    def __init__(self, vector_size=128, window=5, min_count=2, workers=4):
        self.vector_size = vector_size
        self.window = window
        self.min_count = min_count
        self.workers = workers
        self.model = None

    def train(self, corpus: List[str]):
        """corpus: list of code strings. Tokenizes and trains Word2Vec skip-gram."""
        tokenized = [tokenize_code(doc) for doc in corpus]
        print(f"Training Word2Vec on {len(tokenized)} documents...")
        self.model = Word2Vec(
            sentences=tokenized,
            vector_size=self.vector_size,
            window=self.window,
            min_count=self.min_count,
            workers=self.workers,
            sg=1,       # skip-gram
            epochs=10,
        )
        print(f"Vocabulary size: {len(self.model.wv)}")
        print(f"Embedding dimension: {self.vector_size}")

    def encode(self, texts: List[str]) -> np.ndarray:
        """Mean-pool token vectors to get one vector per text."""
        embeddings = []
        for text in texts:
            tokens = tokenize_code(text)
            vecs = [self.model.wv[t] for t in tokens if t in self.model.wv]
            if vecs:
                embeddings.append(np.mean(vecs, axis=0))
            else:
                embeddings.append(np.zeros(self.vector_size))
        return np.array(embeddings, dtype=np.float32)
````

````python
# =====================================================================
# CELL 5: Build Knowledge Base
# =====================================================================
# Train Word2Vec on ALL training code (both buggy and fixed versions),
# so the vocabulary covers both sides of the bug-fix transformation.
# Then embed each training example by its BUGGY code — because at
# inference time, the query is a buggy snippet.

train_data = dataset["train"]

# Combine all code for Word2Vec training vocabulary
all_code = list(train_data["buggy"]) + list(train_data["fixed"])
print(f"Total code snippets for Word2Vec training: {len(all_code)}")

# Train the embedder
embedder = CodeEmbedder(vector_size=128, window=5, min_count=2)
embedder.train(all_code)

# Embed the BUGGY code (this is what we search against at inference time)
buggy_embeddings = embedder.encode(list(train_data["buggy"]))
print(f"Buggy embeddings shape: {buggy_embeddings.shape}")

# Build sklearn NearestNeighbors index
dimension = buggy_embeddings.shape[1]
nn_index = NearestNeighbors(n_neighbors=3, metric='euclidean', algorithm='auto')
nn_index.fit(buggy_embeddings)
print(f"NN index: {len(buggy_embeddings)} vectors, {dimension} dimensions")

# Store metadata (the actual buggy→fixed pairs) for retrieval
kb_metadata = [
    {"buggy": train_data[i]["buggy"], "fixed": train_data[i]["fixed"]}
    for i in range(len(train_data))
]
````

````python
# =====================================================================
# CELL 6: Save Knowledge Base to Disk (Optional — for persistence)
# =====================================================================
KB_DIR = "./rag_refinement_kb"
os.makedirs(KB_DIR, exist_ok=True)

with open(f"{KB_DIR}/nn_index.pkl", "wb") as f:
    pickle.dump(nn_index, f)
embedder.model.save(f"{KB_DIR}/word2vec.model")

with open(f"{KB_DIR}/metadata.pkl", "wb") as f:
    pickle.dump(kb_metadata, f)

config = {
    "embedding_method": "word2vec",
    "vector_size": embedder.vector_size,
    "dimension": dimension,
    "num_examples": len(kb_metadata),
}
with open(f"{KB_DIR}/config.json", "w") as f:
    json.dump(config, f, indent=2)

print(f"Knowledge base saved to {KB_DIR}")
````

````python
# =====================================================================
# CELL 7: Retriever Class
# =====================================================================

class BugFixRetriever:
    """Retrieves similar buggy→fixed pairs from the knowledge base."""

    def __init__(self, kb_dir=KB_DIR):
        """Load a previously saved knowledge base from disk."""
        print(f"Loading KB from: {kb_dir}")

        with open(f"{kb_dir}/config.json", "r") as f:
            self.config = json.load(f)

        with open(f"{kb_dir}/nn_index.pkl", "rb") as f:
            self.index = pickle.load(f)

        with open(f"{kb_dir}/metadata.pkl", "rb") as f:
            self.metadata = pickle.load(f)

        self.embedder = CodeEmbedder(vector_size=self.config["vector_size"])
        self.embedder.model = Word2Vec.load(f"{kb_dir}/word2vec.model")

        print(f"Loaded {self.config['num_examples']} examples, "
              f"{self.config['dimension']}-dim embeddings")

    def retrieve(self, buggy_code: str, top_k: int = 3) -> List[Dict]:
        """Given buggy code, find the top-k most similar buggy→fixed pairs."""
        query_embedding = self.embedder.encode([buggy_code])
        distances, indices = self.index.kneighbors(query_embedding, n_neighbors=top_k)

        results = []
        for dist, idx in zip(distances[0], indices[0]):
            if idx < 0 or idx >= len(self.metadata):
                continue
            item = self.metadata[idx]
            results.append({
                "buggy": item["buggy"],
                "fixed": item["fixed"],
                "similarity": float(1 / (1 + dist)),
            })
        return results


# Initialize the retriever (loads from disk)
retriever = BugFixRetriever(KB_DIR)
````

````python
# =====================================================================
# CELL 8: Test Retrieval
# =====================================================================
test_buggy = dataset["test"][0]["buggy"]
print(f"Query (buggy code):\n{test_buggy[:300]}...\n")

retrieved = retriever.retrieve(test_buggy, top_k=3)
print(f"Retrieved {len(retrieved)} similar bug-fix examples:\n")
for i, r in enumerate(retrieved, 1):
    print(f"  {i}. Similarity: {r['similarity']:.4f}")
    print(f"     Buggy: {r['buggy'][:120]}...")
    print(f"     Fixed: {r['fixed'][:120]}...\n")
````

````python
# =====================================================================
# CELL 9: Load Qwen Generator
# =====================================================================
MODEL_NAME = "Qwen/Qwen2.5-Coder-1.5B-Instruct"

qwen_tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
qwen_model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True,
)

if qwen_tokenizer.pad_token is None:
    qwen_tokenizer.pad_token = qwen_tokenizer.eos_token
qwen_tokenizer.padding_side = "left"

print(f"Loaded {MODEL_NAME} on {qwen_model.device}")
````

````python
# =====================================================================
# CELL 10: Prompt Building and Code Generation
# =====================================================================

def build_bugfix_prompt(buggy_code: str, rag_context: str = "") -> str:
    """Build a Qwen chat prompt for bug fixing.

    Args:
        buggy_code: The buggy Java code to fix.
        rag_context: Optional few-shot context from retrieved examples.

    Returns:
        Formatted prompt string ready for Qwen.
    """
    system = (
        "You are an expert Java developer specializing in bug fixing. "
        "Given buggy Java code, produce the corrected version. "
        "Return ONLY the fixed code, no explanations."
    )
    if rag_context:
        system += " Use the provided bug-fix examples as reference for patterns."

    if rag_context:
        user_content = (
            f"{rag_context}"
            f"Now fix the following buggy Java code:\n"
            f"```java\n{buggy_code}\n```"
        )
    else:
        user_content = (
            f"Fix the following buggy Java code. Return only the corrected code.\n"
            f"```java\n{buggy_code}\n```"
        )

    chat = [
        {"role": "system", "content": system},
        {"role": "user", "content": user_content},
    ]
    return qwen_tokenizer.apply_chat_template(
        chat, tokenize=False, add_generation_prompt=True
    )


def build_rag_context(retrieved_examples: List[Dict]) -> str:
    """Build few-shot context string from retrieved buggy→fixed pairs."""
    if not retrieved_examples:
        return ""

    parts = ["Here are similar Java bug-fix examples for reference:\n"]
    for i, ex in enumerate(retrieved_examples, 1):
        parts.append(f"--- Example {i} ---")
        parts.append(f"Buggy:\n```java\n{ex['buggy']}\n```")
        parts.append(f"Fixed:\n```java\n{ex['fixed']}\n```\n")
    return "\n".join(parts)


def generate_batch(prompts: List[str], max_new_tokens: int = 512) -> List[str]:
    """Generate code for a batch of prompts using Qwen (greedy decoding)."""
    inputs = qwen_tokenizer(
        prompts, return_tensors="pt", padding=True, truncation=True, max_length=4096
    ).to(qwen_model.device)
    input_seq_len = inputs["input_ids"].shape[1]  # padded input length

    with torch.no_grad():
        outputs = qwen_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,  # greedy decoding for deterministic output
            temperature=1.0,  # override model defaults to suppress warnings
            top_p=1.0,
            top_k=50,
            pad_token_id=qwen_tokenizer.pad_token_id,
            eos_token_id=qwen_tokenizer.eos_token_id,
        )

    results = []
    for seq in outputs:
        # Slice off the input portion (padded to same length across batch)
        generated_tokens = seq[input_seq_len:]
        code = qwen_tokenizer.decode(generated_tokens, skip_special_tokens=True).strip()
        results.append(code)
    return results


def generate_code(prompt: str, max_new_tokens: int = 512) -> str:
    """Generate code for a single prompt (convenience wrapper)."""
    return generate_batch([prompt], max_new_tokens)[0]


def extract_code(text: str) -> str:
    """Extract code from markdown code blocks that Qwen may wrap output in."""
    pattern = r'```(?:java)?\s*\n?(.*?)```'
    matches = re.findall(pattern, text, re.DOTALL)
    if matches:
        return matches[0].strip()

    # Fallback: strip leading markdown markers
    for prefix in ["```java", "```"]:
        if text.startswith(prefix):
            return text.replace(prefix, "", 1).strip()

    return text.strip()
````

````python
# =====================================================================
# CELL 11: Quick Sanity Check
# =====================================================================
# Test both zero-shot and RAG prompts on a single example

test_buggy_code = dataset["test"][0]["buggy"]
test_fixed_code = dataset["test"][0]["fixed"]

# Zero-shot
zs_prompt = build_bugfix_prompt(test_buggy_code)
zs_output = generate_code(zs_prompt)
zs_cleaned = extract_code(zs_output)

print("=== ZERO-SHOT ===")
print(f"BUGGY:  {test_buggy_code[:200]}...")
print(f"PRED:   {zs_cleaned[:200]}...")
print(f"LABEL:  {test_fixed_code[:200]}...")

# RAG (3-shot)
retrieved = retriever.retrieve(test_buggy_code, top_k=3)
rag_ctx = build_rag_context(retrieved)
rag_prompt = build_bugfix_prompt(test_buggy_code, rag_context=rag_ctx)
rag_output = generate_code(rag_prompt)
rag_cleaned = extract_code(rag_output)

print("\n=== RAG (3-shot) ===")
print(f"PRED:   {rag_cleaned[:200]}...")
print(f"LABEL:  {test_fixed_code[:200]}...")
````

````python
# =====================================================================
# CELL 12: Full Evaluation — Zero-Shot and RAG (Batched)
# =====================================================================

# Toggle this for quick testing vs full run
DEBUG_MODE = False
NUM_DEBUG_SAMPLES = 10
TOP_K = 3   # number of retrieved examples for RAG
BATCH_SIZE = 16  # RTX 5000 Ada (32GB) — lower to 16 if OOM

test_data = dataset["test"]
if DEBUG_MODE:
    eval_indices = list(range(NUM_DEBUG_SAMPLES))
else:
    eval_indices = list(range(len(test_data)))

num_samples = len(eval_indices)
print(f"Evaluating on {'DEBUG (' + str(num_samples) + ' samples)' if DEBUG_MODE else 'FULL'} test set...")
print(f"Batch size: {BATCH_SIZE}")

# Storage for predictions
zs_preds = []   # zero-shot predictions
rag_preds = []  # RAG predictions
labels = []     # ground truth

# Process in batches
for batch_start in tqdm(range(0, num_samples, BATCH_SIZE), desc="Evaluating"):
    batch_indices = eval_indices[batch_start : batch_start + BATCH_SIZE]

    batch_buggy = [test_data[i]["buggy"] for i in batch_indices]
    batch_fixed = [test_data[i]["fixed"] for i in batch_indices]
    labels.extend(batch_fixed)

    # --- Zero-shot (batched) ---
    zs_prompts = [build_bugfix_prompt(b) for b in batch_buggy]
    zs_outputs = generate_batch(zs_prompts)
    zs_preds.extend([extract_code(o) for o in zs_outputs])

    # --- RAG (batched) ---
    rag_prompts = []
    for b in batch_buggy:
        retrieved = retriever.retrieve(b, top_k=TOP_K)
        rag_ctx = build_rag_context(retrieved)
        rag_prompts.append(build_bugfix_prompt(b, rag_context=rag_ctx))
    rag_outputs = generate_batch(rag_prompts)
    rag_preds.extend([extract_code(o) for o in rag_outputs])

    # Periodic cache clear
    torch.cuda.empty_cache()

print(f"\nGeneration complete. {len(zs_preds)} zero-shot, {len(rag_preds)} RAG predictions.")
````

````python
# =====================================================================
# CELL 13: Print Sample Predictions (Debug)
# =====================================================================
if DEBUG_MODE:
    for i in range(min(5, num_samples)):
        print(f"\n{'='*60}")
        print(f"Example {i}")
        print(f"{'='*60}")
        print(f"BUGGY:    {test_data[i]['buggy'][:200]}...")
        print(f"ZS PRED:  {zs_preds[i][:200]}...")
        print(f"RAG PRED: {rag_preds[i][:200]}...")
        print(f"LABEL:    {labels[i][:200]}...")
````

````python

# =====================================================================
# CELL 14: Compute Metrics — CodeBLEU + Exact Match
# =====================================================================

def evaluate_predictions(predictions, references, label=""):
    """Compute CodeBLEU and Exact Match for a set of predictions."""
    print(f"\n{'='*50}")
    print(f"Results: {label}")
    print(f"{'='*50}")

    # Exact Match
    exact_matches = sum(
        p.strip() == r.strip() for p, r in zip(predictions, references)
    )
    exact_match_acc = exact_matches / len(predictions)
    print(f"Exact Match: {exact_match_acc:.4f} ({exact_matches}/{len(predictions)})")

    # CodeBLEU
    try:
        codebleu_result = calc_codebleu(
            references=[[ref] for ref in references],
            predictions=predictions,
            lang="java",
        )
        print(f"CodeBLEU:       {codebleu_result['codebleu']:.4f}")
        print(f"  - ngram:      {codebleu_result['ngram_match_score']:.4f}")
        print(f"  - weighted:   {codebleu_result['weighted_ngram_match_score']:.4f}")
        print(f"  - syntax:     {codebleu_result['syntax_match_score']:.4f}")
        print(f"  - dataflow:   {codebleu_result['dataflow_match_score']:.4f}")
    except Exception as e:
        print(f"⚠️ CodeBLEU computation failed: {e}")

    return {"exact_match": exact_match_acc}


# Evaluate both settings
zs_metrics = evaluate_predictions(zs_preds, labels, label="Zero-Shot (No RAG)")
rag_metrics = evaluate_predictions(rag_preds, labels, label=f"RAG ({TOP_K}-shot)")

# Side-by-side comparison
print(f"\n{'='*50}")
print("COMPARISON SUMMARY")
print(f"{'='*50}")
print(f"{'Setting':<25} {'Exact Match':>12}")
print(f"{'-'*37}")
print(f"{'Zero-Shot (No RAG)':<25} {zs_metrics['exact_match']:>12.4f}")
print(f"{'RAG (' + str(TOP_K) + '-shot)':<25} {rag_metrics['exact_match']:>12.4f}")
````
